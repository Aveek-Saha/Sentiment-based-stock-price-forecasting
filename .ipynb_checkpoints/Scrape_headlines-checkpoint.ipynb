```python
print(len(res))

    
if len(res) == 0:
    response = urllib.request.urlopen(base_url + "begin_date="+ date +"&end_date=" + date + fail_params)
    html = response.read()
    # encoding = response.info().get_content_charset('utf-8')
    jres = json.loads(html)

    res = jres['response']['docs']
    
    if len(res) == 0:
        response = urllib.request.urlopen(base_url + "begin_date="+ date +"&end_date=" + date + double_fail_params)
        html = response.read()
        # encoding = response.info().get_content_charset('utf-8')
        jres = json.loads(html)

        res = jres['response']['docs']
    
    


for x in res:
    print(x['headline']['main'])
    print(x['snippet'])
    print()
```

```python
final_news = []
count = 0

for dat in df_stock['Date']:
    date = "".join(dat.split('-'))
    
    response = urllib.request.urlopen(base_url + "begin_date="+ date +"&end_date=" + date +params)
    html = response.read()
    jres = json.loads(html)

    res = jres['response']['docs']
    
    time.sleep(5)
    
    print("Length",len(res))

    
    if len(res) == 0:
        response = urllib.request.urlopen(base_url + "begin_date="+ date +"&end_date=" + date + fail_params)
        html = response.read()
        # encoding = response.info().get_content_charset('utf-8')
        jres = json.loads(html)

        res = jres['response']['docs']
        time.sleep(4)
        

        if len(res) == 0:
            response = urllib.request.urlopen(base_url + "begin_date="+ date +"&end_date=" + date + double_fail_params)
            html = response.read()
            # encoding = response.info().get_content_charset('utf-8')
            jres = json.loads(html)

            res = jres['response']['docs']
            time.sleep(4)
            


    news = []
    count +=1
    print("Done ",count)
    print()

    for x in res:
        news.append([x['headline']['main'], x['snippet']])
    final_news.append([date,news])
        
```

```python
final_news
```

```python
new_arr = []

for date, news in final_news:
    total_n=""
    for head, snip in news:
        total_n += head+" "+snip+'|'
#     print(date, total_n)
    new_arr.append([date, total_n])
#     print()
```

```python
df_news = pd.DataFrame(new_arr, columns=['Date', 'News'])
```

```python
df_news.to_csv('datasets/news.csv', index=False)
```
