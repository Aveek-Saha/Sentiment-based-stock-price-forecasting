# Get tweets using the twitterscraper library

```python
import csv
import pandas as pd
import numpy as np
import datetime as dt
from datetime import timedelta
import unicodedata

from twitterscraper import query_tweets
```

```python
df_stock = pd.read_csv('./datasets/AAPL.csv')
df_stock.head()
```

```python
df_stock['Date'][0].split('-')
```

### Write the scraped tweets to a file

```python
row = ['Date', 'Tweets']

with open('./datasets/tweets.csv', 'a') as csvFile:
    writer = csv.writer(csvFile)
    writer.writerow(row)

csvFile.close()
```

```python
all_tweets = []

for date in df_stock['Date']:
    darr = date.split('-')
    start = dt.date(int(darr[0]), int(darr[1]), int(darr[2]))
    end = start + timedelta(days=1)
    
    tweets = []

    for tweet in query_tweets("apple OR iphone OR ipad OR macbook", 10, begindate=start, enddate=end, lang='en'):
        tweets.append(unicodedata.normalize('NFKD', tweet.text).encode('ascii', 'ignore').decode('utf-8'))
        
    
    row = [date, "|".join(tweets)]

    with open('./datasets/tweets.csv', 'a') as csvFile:
        writer = csv.writer(csvFile)
        writer.writerow(row)

    csvFile.close()
    print(date)

```
